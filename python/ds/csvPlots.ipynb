```python
import threading
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv('shows.csv')
print(data_set)
```

### Count for each categoryDescription

```python
def plot_category_description():
    data_set['categoryDescription'].value_counts().plot(kind='bar')
    plt.show()
```

### Count for each IPgeo

```python
def plot_ip_geo():
    data_set['IPgeo'].value_counts().plot(kind='bar')
    plt.show()
```

## Threads creation

```python
def thread_function(id):
    if id == 0:
        print('Thread 0')
        plot_category_description()
    elif id == 1:
        plot_ip_geo()

num_of_threads = 2
threads = []
for i in range(num_of_threads):
    threads.append(
        threading.Thread(target=thread_function, args=(i,))
    )

for thread in threads:
    thread.start()

for thread in threads:
    thread.join()
```
